# Analyze Trends Over Time


```python
#Ensure 'release_year' is in numeric format
df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce')
```

```python
# Drop rows with missing release_year
df = df.dropna(subset=['release_year'])
```

```python
#Group by release year and calculate the average duration
yearly_avg_duration = df.groupby('release_year')['duration'].mean()
```

```python
#Plot the trends over time
plt.figure(figsize=(12, 6))

plt.plot(yearly_avg_duration.index, yearly_avg_duration.values, marker='o', linestyle='-', color='r')

plt.xlabel('Release Year')
plt.ylabel('Average Duration (minutes)')
plt.title('Average Movie/Episode Duration Over Time')
plt.grid(True)
plt.show()
```

# Analyze by Genre or Content Type

```python
plt.figure(figsize=(12, 8))

# Create a box plot of durations by type
sns.boxplot(x='type', y='duration', data=df, palette='viridis')

plt.xlabel('Content Type')
plt.ylabel('Duration (minutes)')
plt.title('Duration Distribution by Content Type')
plt.show()
```
